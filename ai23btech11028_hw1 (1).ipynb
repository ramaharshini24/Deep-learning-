```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50, 50, 100) #(start, stop, dist) this x is an array that contians the values of x
g = np.exp(x)/ (1 + np.exp(x))
plt.plot(x, g)
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.show()
```

```python
#3(1)
import numpy as np
import matplotlib.pyplot as plt

# Define a grid
x = np.linspace(-1.5, 1.5, 400)
y = np.linspace(-1.5, 1.5, 400)
X, Y = np.meshgrid(x, y)

# Norms
L1 = np.abs(X) + np.abs(Y)
L2 = np.sqrt(X**2 + Y**2)

# 1. L1 norm ball open
plt.contourf(X, Y, L1, levels=[0, 1], colors=['lightblue'], alpha=0.5)
plt.contour(X, Y, L1, levels=[1], colors='blue', linestyles='dashed')
plt.title("L1--open ball")
plt.gca().set_aspect('equal')
plt.show()

# 2. L1 norm ball closed
plt.contourf(X, Y, L1, levels=[0, 1], colors=['lightgreen'], alpha=0.5)
plt.contour(X, Y, L1, levels=[1], colors='green', linewidths=2)
plt.title("L1--closed ball")
plt.gca().set_aspect('equal')
plt.show()

# 3. L2 norm ball open
plt.contourf(X, Y, L2, levels=[0, 1], colors=['lightcoral'], alpha=0.5)
plt.contour(X, Y, L2, levels=[1], colors='red', linestyles='dashed')
plt.title("L2--open ball")
plt.gca().set_aspect('equal')
plt.show()

# 4. L2 norm ball closed
plt.contourf(X, Y, L2, levels=[0, 1], colors=['lightyellow'], alpha=0.5)
plt.contour(X, Y, L2, levels=[1], colors='orange', linewidths=2)
plt.title("L2--closed ball")
plt.gca().set_aspect('equal')
plt.show()
```
